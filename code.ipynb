```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\HP\leto_24\cartogramma\оборот розничной торговли на душу населения(ДФО).xlsx')
df
```

```python
df.columns
```

```python
df = df.rename(columns = {'Unnamed: 0' : 'region',
                         2019: '2019',
                         2020: '2020',
                         2021: '2021',
                         2022: '2022',
                         2023: '2023'})
```

```python
df_copy = df.dropna(axis = 0)
```

```python
import geopandas as gpd
```

```python
file = gpd.read_file(r'C:\Users\HP\leto_24\cartogramma\region.geojson')
file
```

```python
df_geo = file[['wikipedia', 'geometry']]
df_geo = df_geo.dropna(axis = 0)
df_geo
```

```python
df_geo['wikipedia'] = list(map(lambda x: x.split()[0] , df_geo['wikipedia'].str.replace('ru:', '', regex=False).str.lower().str.strip()))
df_geo
```

```python
df_copy
```

```python
df_copy['region'] = list(map(lambda x: x.split()[0], df_copy['region'].str.replace('республика бурятия', 'бурятия').str.replace('чукотскийао', 'чукотский').str.lower().str.strip()))
df_copy
```

```python
prim = df_geo.merge(df_copy, how='left', left_on = 'wikipedia', right_on = 'region')
prim
```

```python

print(max(prim['2023']), min(prim['2023']))
print((max(prim['2023']) - min(prim['2023'])) / 4)
```

```python
bin = [0, 100, 250, 400, 700]
labels = ['0-99', '100-249', '250-399', 'более 400']

prim['Категория'] = pd.cut(prim['2023'], bins=bin, labels=labels)

if prim['Категория'].isna().any():
    prim['Категория'] = prim['Категория'].cat.add_categories(['Нет данных'])
    prim['Категория'].fillna('Нет данных', inplace=True)
prim
```

```python
prim.drop(7, inplace = True)
prim
```

```python
color_map = {
    '0-99': '#A9D3FD',
    '100-249': '#7DBBFC',
    '250-399': '#346FC2',
    'более 400': '#363194',
    'Нет данных': '#BFBFBF'
}
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))  # Установка размера графика
prim.plot(color=prim['Категория'].map(color_map),
          edgecolor='white',
          linewidth=1,
          ax=ax)  # Добавление оси

# Заголовок и отключение осей
plt.title('Оборот розничной торговли в Дальневосточном федеральном округе на 2023г, млрд')
plt.axis('off')

# Добавление легенды
handles = [plt.Line2D([0], [0], marker='o', color='w', label=label,
                       markerfacecolor=color_map[label], markersize=10) for label in prim['Категория'].unique()]
plt.legend(handles=handles, loc='lower right')

# Показ графика
plt.show()
```
